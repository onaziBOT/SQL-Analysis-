## Hour with the most sales in diffrent citys Plot

```python
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv("Hour by City.csv")
df = dataset
```

```python
plt.figure(figsize=(10,6))
for city in df["City"].unique():
    city_data = df[df["City"] == city]
    plt.plot(city_data["Hour"], city_data["count_per_hour"], label=city)

plt.title('Hourly Count per City')
plt.xlabel('Hour')
plt.ylabel('Count per Hour')
plt.legend()
plt.grid(True)
plt.show()
```
